# Internship Task 1: Data Preparation and Cleaning
## Project Goal
The objective of this task is to take a raw dataset and clean it to ensure data integrity and readiness for subsequent analytical tasks. This involves identifying structural issues, handling missing information, removing redundancies, and establishing proper data formatting.

## Key Requirements & Workflow
1. **Identify Missing or Null Values:** Scan the dataset to locate gaps or unrecorded data points across all features.
2. **Remove Redundancy:** Detect and drop exact duplicate records to avoid skewed analytical results.
3. **Correct Data Formats:** Standardize data types, ensuring dates are recognizable as timestamps and numbers are parsed accurately without text interference.
## Technical Stack & Skills Demonstrated
* **Environment:** Python (Jupyter Notebook)
* **Libraries Used:** `pandas`, `numpy`
* **Core Competencies:** Data Cleaning, Data Type Casting, Handling Missing Data (Imputation/Flagging)

```python
import pandas as pd

df = pd.read_csv("Dataset for Data Analytics - Sheet1.csv")

print("--- Missing Values Count ---")
print(df.isnull().sum())
print("\n" + "="*40 + "\n")

duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")
```

```python
df_cleaned = df.drop_duplicates()

df['CouponCode'] = df['CouponCode'].fillna('None')

df['Date'] = pd.to_datetime(df['Date'])

df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
df['TotalPrice'] = pd.to_numeric(df['TotalPrice'], errors='coerce')

print("Data cleaning steps applied successfully!")
```

```python
print("--- New Missing Values Count ---")
print(df.isnull().sum())
print("\n" + "="*40 + "\n")

print("--- Updated Column Data Types ---")
df.info()

df.to_csv("Cleaned_Dataset_Task1.csv", index=False)
print("\nCleaned dataset saved as 'Cleaned_Dataset_Task1.csv'!")
```
